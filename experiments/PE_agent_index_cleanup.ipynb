```python
import copy
import os
import os.path as osp

import pandas as pd

results_dir = "/home/jonathon/baposgmcp_results/PursuitEvasion16x16-v0/exp_36689_seed1/klr_initseed1_numseeds1_2023-01-25_22-44-13racxjj2_"
```

```python
result_filepaths = []
for dirpath, dirnames, filenames in os.walk(results_dir):
    for fname in filenames:
        if fname.endswith("csv"):
            result_filepaths.append(os.path.join(dirpath, fname))

result_filepaths.sort()


def fix_file(fpath):
    df = pd.read_csv(fpath)
    
    policy_ids = df["policy_id"].unique()
    changed = False
    for pi_id in policy_ids:
        if "_i0" in pi_id or "_i1" in pi_id:
            continue
            
        pi_df = df[df["policy_id"] == pi_id]
        agent_id = pi_df["agent_id"].unique()
        for i in agent_id:        
            tokens = pi_id.split("_")
            if tokens[0] == "ucbmcp-random":
                tokens.insert(1, f"i{i}")
            else:
                tokens.insert(2, f"i{i}")
            new_pi_id = "_".join(tokens)
            
        df.loc[(df["policy_id"] == pi_id) & (df["agent_id"] == i)] = new_pi_id
        changed = True
            
    if not changed:
        return
    
    print("changed", fpath)
    
    df.to_csv(fpath, index=False)

for fpath in result_filepaths:
    fix_file(fpath)
```

```python
for fpath in result_filepaths:
    df = pd.read_csv(fpath)
    print("\n", fpath)
    exp_ids = df["exp_id"].unique().tolist()
    exp_ids.sort()
    print("Exp ID:", exp_ids)
```
